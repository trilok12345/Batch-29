# Lists And Arrays

```python
import numpy as np
list1 = [1,2,3]
array1 = np.array(list1)
print(array1)
```

```python
type(list1)
```

```python
type(array1)
```

```python
list2 = [5,6,7]
array2 = np.array(list2)
print(array2)
```

```python

array3 = array1 + array2
print(array3)
```

```python
print('Product of array1 and array2 is :',array1*array2)
print('Division of array2 by array1 is :',array2/array1)
print('array2 raised to the power of array1 is :',array2**array1)
```

```python
# Sin Function
print('Sine : ',np.sin(array1))

# Logarithm
print('Natural Logarithm : ',np.log(array1))
print('Base-10 Logarithm : ',np.log10(array1))
print('Base-1 Logarithm : ',np.log2(array1))

# Exponential
print('Exponential :',np.exp(array1))
```

# Datasets

```python
fnd = ['Jayanth','sanjay','mahendra','maruthi','akhil','deepu','sweety','nikhil','abhi','shireesha']
age = [17,19,19,22,24,49,16,20,18,25]
weight = [52,45,50,55,60,40,45,56,39,48]
height = [140,135,170,156,179,168,135,145,160,132]
```

```python
import matplotlib.pyplot as mat
mat.scatter(weight,height)
mat.title('Relationship between Height and Weight')
mat.ylabel('height in cm')
mat.xlabel('Weight in kg')
mat.show()
```

```python
mat.scatter(age,height)
mat.title('Relationship between Height and Age')
mat.ylabel('height in cm')
mat.xlabel('Age')
mat.show()
```

# Bargraph

```python
mat.figure(figsize=(12,5))
mat.title('Friends Weight in kg',fontsize=10,fontstyle='italic')
# Main Plot function
mat.bar(x=fnd,height=weight, width=0.6, color='red',edgecolor='k',alpha=0.6)
mat.xlabel("F R I E N D S",fontsize=12)
mat.ylabel('Weight in kg',fontsize=12)
```
